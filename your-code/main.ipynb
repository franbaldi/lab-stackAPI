#### Question 1: Find the questions and answers of last month.

```python
SITE = StackAPI('stackoverflow')
questions = SITE.fetch('questions', fromdate=1584226800, todate=1586901600)
questions
```

```python
# Creating list for questions title's 

questions_titles = []
for i in range(500):
    questions_titles.append(questions['items'][i]["title"])
    
questions_titles
```

```python
# Getting the JSON data for answers
SITE = StackAPI('stackoverflow')
answers = SITE.fetch('answers', fromdate=1584226800, todate=1586901600)
answers
print(json.dumps(answers, indent=6, sort_keys=True))
```

```python
# Getting all the answers display names in a list
answers_display_names = []

for i in range(500):
    for x in range(500):
        answers_display_names.append(answers['items'][i]["owner"][i]["display_name"])
    
answers_display_names
```

#### Question 2: Find the most voted question today with at least a score of 5 and tagged with 'python'.
